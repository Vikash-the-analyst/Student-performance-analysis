```python
fig,(ax1, ax2)= plt.subplots(1,2, figsize=(10,5))
fig.suptitle("math score distribution analysis")
sns.kdeplot(data= df, x = "math score", color= "orange", fill= True, ax=ax1)
ax1.title.set_text("KDE")
sns.ecdfplot(data= df, x= "math score", ax=ax2)
ax2.title.set_text("CDF")
plt.show()
```

```python
fig,(ax1, ax2)= plt.subplots(1,2, figsize=(10,5))
fig.suptitle("reading score distribution analysis")
sns.kdeplot(data= df, x = "reading score", color= "orange", fill= True, ax=ax1)
ax1.title.set_text("KDE")
sns.ecdfplot(data= df, x= "reading score", ax=ax2)
ax2.title.set_text("CDF")
plt.show()
```

```python
fig,(ax1, ax2)= plt.subplots(1,2, figsize=(10,5))
fig.suptitle("writing score distribution analysis")
sns.kdeplot(data= df, x = "writing score", color= "orange", fill= True, ax=ax1)
ax1.title.set_text("KDE")
sns.ecdfplot(data= df, x= "writing score", ax=ax2)
ax2.title.set_text("CDF")
plt.show()
```

# observation :-
### The distribution of all three shows they are left/negatively skewed.
### The average scores for writing, reading, and math are around 65 to 70.
### Most of the students have scored above 40 marks.
### The scores for writing and reading have less variation than the scores for math, which have more outliers.


```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(12,6))
fig.suptitle("Performance analysis by gender")
sns.boxplot(data= df, x="gender", y= "math score", ax= ax1)
sns.boxplot(data= df,x="gender", y="reading score", ax= ax2)
sns.boxplot(data= df, x="gender", y="writing score", ax=ax3)
plt.show()
```

# observation :-
### In math Female student score zero, But male students perform good.
### In Reading and writing female students perform better than male students.

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(12,6))
fig.suptitle("Performance analysis by test preparation")
sns.boxplot(data= df, x="test preparation course", y= "math score", ax= ax1)
sns.boxplot(data= df,x="test preparation course", y="reading score", ax= ax2)
sns.boxplot(data= df, x="test preparation course", y="writing score", ax=ax3)
plt.show()
```

# observation :-
### Students who completed their test preparation has performed better than who does not complete test prepartion.

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(12,6))
fig.suptitle("Performance analysis by race/ethnicity")
sns.boxplot(data= df, x="race/ethnicity", y= "math score", ax= ax1)
sns.boxplot(data= df,x="race/ethnicity", y="reading score", ax= ax2)
sns.boxplot(data= df, x="race/ethnicity", y="writing score", ax=ax3)
plt.show()
```

## Observation:-
### students belongs from group E are performed very good.
### students from group A has very poor performance.

```python
# calculating total score for better understanding

df["total score"]= df["math score"]+df["reading score"]+df["writing score"]
df["average score"]= df["total score"]/3
```

```python
fig,(ax1, ax2) =plt.subplots(1,2, figsize=(10,5))
sns.histplot( data=df, x="average score", color= "green", kde=True, ax= ax1)
plt.subplot(121)
sns.histplot( data=df, x="average score", color= "green", hue="gender", kde=True, ax= ax2)
plt.subplot(122)
plt.show()
```

### Observation:-
### female students perform better than male students.

```python
fig,(ax1, ax2, ax3) = plt.subplots(1,3,figsize=(15,7))
plt.subplot(131)
sns.histplot(data=df,x='average score',kde=True,hue='lunch',ax=ax1)
plt.subplot(132)
sns.histplot(data=df[df.gender=='female'],x='average score',kde=True,hue='lunch', ax=ax2)
ax2.title.set_text("female")
plt.subplot(133)
sns.histplot(data=df[df.gender=='male'],x='average score',kde=True,hue='lunch', ax=ax3)
ax3.title.set_text("male")
plt.show()
```

# observation :-
### standard lunch help students to perform well in exam.

```python

fig,(ax1, ax2, ax3) = plt.subplots(1,3,figsize=(15,7))
plt.subplot(131)
sns.histplot(data=df,x='average score',kde=True,hue='parental level of education', ax=ax1)
plt.subplot(132)
sns.histplot(data=df[df.gender=='female'],x='average score',kde=True,hue='parental level of education', ax=ax2)
ax2.title.set_text("female")
plt.subplot(133)
sns.histplot(data=df[df.gender=='male'],x='average score',kde=True,hue='parental level of education', ax=ax3)
ax3.title.set_text("male")
plt.show()
```

# observation :-
### Parents education does make impact on students performance.
### In 2nd Graph we see that female performance does not make any effects on their performance.
### In 3rd graph we see that male candidated have perform well. 

```python
df1=df[["math score", "reading score", "writing score", "average score","total score"]]
sns.heatmap(df1.corr(),annot = True)
plt.show()
```
